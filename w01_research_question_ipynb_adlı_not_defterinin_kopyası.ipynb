**The question:** decision, action, cost of a wrong call

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd


csv_url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(csv_url)

# 1. Calculate the overall decline rate
decline_rate = (df["trend_direction"] == "down").mean()

# 2. Calculate stale & visible pages (high risk, high opportunity)
stale_and_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()

print("--- Data Proof for Lane Selection ---")
print(f"1. Overall Decline Rate: {decline_rate*100:.1f}% of the pages in the dataset are already showing a downward trend.")
print(f"2. Unattended Value: There are {stale_and_visible} pages that have >500 impressions but haven't been updated in over 6 months, representing immediate refresh opportunities.")
```
