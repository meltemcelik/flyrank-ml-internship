## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Binary Classification and Ranking.
First, the model performs binary classification to predict if a page belongs to the "declining" class. Then, it outputs a probability score, turning the task into a ranking problem where pages are sorted by their likelihood of decay to prioritize the review queue.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

For this starter dataset, the proxy label is is_declining_label, derived from the condition trend_direction == "down". While a perfect target would look at future traffic drops over a forward-looking window, this current-state proxy helps establish the baseline logic for identifying decay.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Precision@50 (or Precision@20).
Overall accuracy is not the main goal here because an SEO team has limited capacity; they cannot refresh thousands of pages at once. What matters is the top of the queue. Precision@50 measures exactly what we care about: out of the top 50 pages the model recommends fixing first, how many are actually declining?

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (e.g., "refresh pages older than 180 days with >500 impressions") is rigid and creates too many false positives. It treats all pages the same. Machine learning beats this because it can weigh multiple non-linear signals simultaneously—such as the relationship between content age, exact ranking position, intent type, and impression volume—to find subtle decay patterns that a simple IF/ELSE statement would completely miss.
